```python
def cal_grade(avg):
    if avg>=90:
        return "A"
    elif avg>=75:
        return "B"
    elif avg>=55:
        return "C"
    elif avg>=45:
        return "D"
    else:
        return "Fail"
sub=int(input("Enter the Number of subjects:"))
total=0
for i in range(sub):
    marks=int(input(f"Enter the marks for subject {i+1}: "))
    total +=marks
avg= round(total/sub,2)
grade=cal_grade(avg)
print("Total:",total)
print("Average:",avg)
print("Grade:",grade)
```
